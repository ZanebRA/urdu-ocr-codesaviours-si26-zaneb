```python
# Install required libraries
!pip -q install pillow arabic-reshaper python-bidi
!apt-get -qq update
!apt-get -qq install -y fonts-noto-core

# Import libraries

from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
import os
# Output folder (Google Drive)

output_folder = "/content/drive/MyDrive/data/data/raw/synthetic"
os.makedirs(output_folder, exist_ok=True)

# Urdu text

urdu_texts = [
    "پاکستان کی ثقافت بہت خوبصورت ہے",
    "ہمیشہ سچ بولنا چاہیے",
    "صاف ماحول صحت کے لیے ضروری ہے",
    "وقت کی قدر کامیابی کی کنجی ہے",
    "محنت انسان کو کامیاب بناتی ہے",
    "علم حاصل کرنا ہر مسلمان پر فرض ہے",
    "درخت لگانا صدقہ جاریہ ہے",
    "صحت مند جسم میں صحت مند دماغ ہوتا ہے",
    "اتحاد میں طاقت ہے",
    "دیانت داری بہترین پالیسی ہے",
    "امید کبھی نہیں چھوڑنی چاہیے",
    "کتابیں انسان کی بہترین دوست ہیں",
    "ہر دن ایک نئی شروعات ہے",
    "پاکستان کے لوگ مہمان نواز ہیں",
    "تعلیم قوموں کی ترقی کا راز ہے",
    "دوسروں کی مدد کرنا اچھی عادت ہے",
    "امن اور محبت سب کے لیے ضروری ہیں",
    "خواب وہی پورے ہوتے ہیں جن کے لیے محنت کی جائے",
    "کامیابی مسلسل کوشش سے ملتی ہے",
    "اپنے والدین کا احترام کریں"
]

# Load Urdu Font
font_path = "/usr/share/fonts/truetype/noto/NotoNaskhArabic-Regular.ttf"
font = ImageFont.truetype(font_path, 48)

# Generate Images
for i, text in enumerate(urdu_texts, start=1):

    reshaped_text = arabic_reshaper.reshape(text)

    img = Image.new("RGB", (700, 180), "white")
    draw = ImageDraw.Draw(img)

    draw.text(
        (630, 70),
        reshaped_text,
        font=font,
        fill="black",
        anchor="ra"
    )

    img.save(f"{output_folder}/synthetic_{i}.png")

    print(f"Generated: synthetic_{i}.png -> {text}")

print("\n 20 synthetic Urdu images generated successfully!")
print(f" Images saved in: {output_folder}")
```

```python
!pip install -q transformers torch pillow pandas
import torch
from torch.utils.data import Dataset
from transformers import TrOCRProcessor
from PIL import Image
import pandas as pd
class UrduOCRDataset(Dataset):

    def __init__(self, csv_path, processor):
        self.data = pd.read_csv(csv_path)
        self.processor = processor
        print(f"Dataset loaded: {len(self.data)} samples")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):

        row = self.data.iloc[idx]

        # Image path
        image_path = "/content/drive/MyDrive/data/" + row["image"]

        # Load image
        image = Image.open(image_path).convert("RGB")

        # Process image
        encoding = self.processor(image, return_tensors="pt")
        pixel_values = encoding.pixel_values.squeeze()

        # Process text
        labels = self.processor.tokenizer(
            row["text"],
            padding="max_length",
            max_length=128
        ).input_ids

        labels = torch.tensor(labels)

        return {
            "pixel_values": pixel_values,
            "labels": labels
        }

print("✅ Dataset class created successfully!")
```

```python
from transformers import TrOCRProcessor

processor = TrOCRProcessor.from_pretrained(
    "microsoft/trocr-base-printed",
    use_fast=False
)

dataset = UrduOCRDataset(
    "/content/drive/MyDrive/data/data/labels.csv",
    processor
)

print("Total samples:", len(dataset))

sample = dataset[0]

print("Pixel Values Shape:", sample["pixel_values"].shape)
print("Labels Shape:", sample["labels"].shape)
```

```python
from torch.utils.data import random_split

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size

train_dataset, test_dataset = random_split(
    dataset,
    [train_size, test_size]
)

print("Training samples:", len(train_dataset))
print("Testing samples:", len(test_dataset))
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=4,
    shuffle=False
)

print("Train DataLoader created successfully!")
print("Test DataLoader created successfully!")
```

```python
def preprocess_image(image_path, save_path):
    # Load image
    img = cv2.imread(image_path)

    if img is None:
        print(f"Could not load: {image_path}")
        return

    # Step 1: Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Step 2: Resize image
    resized = cv2.resize(gray, (512, 128))

    # Step 3: Remove noise
    denoised = cv2.fastNlMeansDenoising(resized, h=10)

    # Step 4: Convert to black and white
    _, binary = cv2.threshold(denoised, 127, 255, cv2.THRESH_BINARY)

    # Save processed image
    cv2.imwrite(save_path, binary)

    return binary

# Create output folder
os.makedirs("data/processed", exist_ok=True)

print("Preprocessing function ready!")
```

```python
import glob
# Path to your raw images in Google Drive
raw_folder = "/content/drive/MyDrive/data/data/raw"

# Find all JPG and PNG images in all subfolders
all_images = glob.glob(raw_folder + "/**/*.jpg", recursive=True)
all_images += glob.glob(raw_folder + "/**/*.png", recursive=True)

print(f"Found {len(all_images)} images to process")

processed_count = 0

for img_path in all_images:
    filename = os.path.basename(img_path)
    save_path = f"data/processed/{filename}"

    result = preprocess_image(img_path, save_path)

    if result is not None:
        processed_count += 1

print(f"Done! Processed {processed_count} images")
print("Check data/processed/ folder")
```

```python
import glob
import os

# Path to raw images
raw_folder = "/content/drive/MyDrive/data/data/raw"

# Find all JPG and PNG images
all_images = glob.glob(raw_folder + "/**/*.jpg", recursive=True)
all_images += glob.glob(raw_folder + "/**/*.png", recursive=True)

print(f"Found {len(all_images)} images to process")

processed_count = 0

for img_path in all_images:

    # Folder name
    folder_name = os.path.basename(os.path.dirname(img_path))

    # Original filename
    filename = os.path.basename(img_path)

    # Unique filename
    new_filename = f"{folder_name}_{filename}"

    # Save path
    save_path = os.path.join("data/processed", new_filename)

    result = preprocess_image(img_path, save_path)

    if result is not None:
        processed_count += 1

print(f"Done! Processed {processed_count} images")
print("Check data/processed/ folder")
```
